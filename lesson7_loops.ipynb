```python
import random
```

```python
rn = random.randint(1,10)
chance = int(input('please guess the number'))
if rn == chance:
    print('you guessed a random number')
else:
    print('try again')
```

```python
rn = random.randint(1,100)
for i in range(10):
    chance = int(input(f'please guess the number, chance {i+1}: '))
    if rn > chance:
        print('number sould be bigger')
    elif rn < chance:
        print('number should be smaller')
    else:
        print('you guessed it right ')
        break
```
